```python
"""the data consist of 8807 rows and 12 columns
show_id,typw,title,release_year, listed_in,description doesnt consist of any missing values 
duration,rating,date_added 2 to 3 missing values 
country,cast,director have 10 and more than 10percentage of data missing 
""""
```

```python
df.columns
```

```python
df = df[['type','title','director','cast','country','date_added','release_year','rating','duration','listed_in']]
```

```python
df
```

```python
#checking number of missing values 
df.isnull().sum()
```

```python
""" based on the above observation on the missing value table we can fill the values in the director,cast,country column as it 2634,825,831 values missing
but in case of date_added,rating,duration they very less missing values so we can drop them
"""
```

```python
# used to fill the missing values as they are more that 10 percent of data missing in this column
df.loc[:, 'director'] = df['director'].fillna('Unknown')
# used to drop data as the data missing is less 10 percent in these columns
df = df.dropna(subset=['cast', 'country', 'date_added', 'rating', 'duration'])
```

```python
#verifying again
df.isnull().sum()
```

```python
df.nunique()
```

```python
#remove duplicates
df.drop_duplicates()
```

```python
df.nunique()
```

```python
df.columns
```

```python
#fix columns names
df.columns = df.columns.str.strip().str.lower().str.replace(" ","_")
```

```python
df['date_added']
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
```

```python
#EDA
```

```python
#visualization
```

```python
#content type distribution
df['type'].value_counts().plot(kind='bar',color=['skyblue','orange'])
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='barh',color='teal')
plt.title('Top 10 Countries Producting Most content')
plt.xlabel('Number of titles')
plt.ylabel('Country')
plt.show()
```

```python
#content growth by year
df["year_added"] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line',marker='o')
plt.title('Netflix content Added by Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python
#most common rating
df['rating'].value_counts().head(10).plot(kind='bar',color='purple')
plt.title('Most common content ratings on netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
#top10 generes
df['listed_in'].value_counts().head(10).plot(kind='barh',color='salmon')
plt.title('Top 10 generes on netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
#movie duration distribution 
movie_df = df[df['type'] == 'Movie']
movie_df['duration_mins'] = movie_df['duration'].str.replace(' min','').astype(float)
movie_df['duration_mins'].plot(kind='hist',bins=20,color='lightgreen',edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()
```
